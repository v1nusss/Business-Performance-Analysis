```python
# -*- coding: utf-8 -*-
"""Business Performance Analysis"""

# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the dataset
file_path = '/content/amazon.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Display basic information about the dataset
print("=== Step 2: Dataset Information ===")
print(data.info())
print("\n=== First 5 Rows of the Dataset ===")
data.head()

```

```python
# Step 3: Data Cleaning - Convert price and rating_count to numeric
data['discounted_price'] = data['discounted_price'].replace('₹', '', regex=True).str.replace(',', '').astype(float)
data['actual_price'] = data['actual_price'].replace('₹', '', regex=True).str.replace(',', '').astype(float)
data['discount_percentage'] = data['discount_percentage'].replace('%', '', regex=True).astype(float)

# Handle NaN values in 'rating_count' before conversion
data['rating_count'] = data['rating_count'].fillna(0).replace(',', '', regex=True).astype(int)

# Ensure the 'rating' column is numeric
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')  # Convert, setting non-numeric values to NaN

# Handle missing or invalid ratings
data['rating'] = data['rating'].fillna(data['rating'].mean())  # Replace NaN ratings with the column mean

print("\n=== Step 3: Cleaned Numeric Columns ===")
data[['discounted_price', 'actual_price', 'discount_percentage', 'rating_count', 'rating']].head()

```

```python
# Step 4: Adding Derived Metrics for Performance Analysis
data['profit_margin'] = data['actual_price'] - data['discounted_price']  # Calculate profit margin
data['total_rating_value'] = data['rating'] * data['rating_count']  # Total cumulative rating

print("\n=== Step 4: Added Derived Metrics ===")
data[['product_name', 'profit_margin', 'total_rating_value']].head()

```

```python
# Step 5: Aggregating Data for Analysis
category_summary = data.groupby('category').agg({
    'discounted_price': 'sum',
    'actual_price': 'sum',
    'profit_margin': 'sum',
    'rating': 'mean',
    'rating_count': 'sum',
    'total_rating_value': 'sum'
}).reset_index()


```

```python
print("\n=== Step 5: Aggregated Category Summary ===")
print(category_summary)

# Step 6: Adding Additional Insights
# Handle missing or zero values in 'rating_count' and 'total_rating_value'
category_summary['total_rating_value'] = category_summary['total_rating_value'].fillna(0)
category_summary['rating_count'] = category_summary['rating_count'].replace(0, pd.NA).fillna(1)

# Calculate additional metrics
category_summary['avg_discounted_price'] = category_summary['discounted_price'] / category_summary['rating_count']
category_summary['avg_actual_price'] = category_summary['actual_price'] / category_summary['rating_count']
category_summary['avg_rating'] = category_summary['total_rating_value'] / category_summary['rating_count']

print("\n=== Step 6: Category Summary with Additional Insights ===")
category_summary[['category', 'avg_discounted_price', 'avg_actual_price', 'avg_rating']]

```

```python
# Step 7: Save Cleaned Data for Tableau
output_file_path = '/content/category_summary.csv'  # Replace with your desired path
category_summary.to_csv(output_file_path, index=False)
print(f"\n=== Step 7: Cleaned Data Saved to {output_file_path} ===")

```

```python
# Step 8: Visualize Key Insights (Optional)
plt.figure(figsize=(10, 6))
category_summary.plot(kind='bar', x='category', y='profit_margin', title='Total Profit Margin by Category', legend=False)
plt.ylabel('Profit Margin')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
